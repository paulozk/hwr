```python
from keras.models import Sequential, model_from_json, Model
from keras.layers import Dense, Dropout, Bidirectional, LSTM, CuDNNLSTM
from keras import backend as K
from keras.optimizers import RMSprop
import keras.callbacks

import numpy as np
import os
import cv2
import matplotlib.pyplot as plt

from create_batches import *


#def load_batcjes():
data_dim = 32
timesteps = 32
num_classes = 27
#pixels = 1875


model = Sequential()

model.add(LSTM(300, input_shape=(timesteps, data_dim)))
#model.add(LSTM(500, input_shape=(timesteps, data_dim)))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='sigmoid'))

get_layer_output = K.function([model.layers[0].input], [model.layers[0].output])




rmsprop = RMSprop(lr=0.005)
model.compile(loss='binary_crossentropy',
              optimizer=rmsprop,
              metrics=['accuracy'])

path = "dataset4"
current_path = os.getcwd()
path = os.path.join(current_path, path)


```

```python
def test_lstm():
    correct = 0
    
    labels = {0: "alef", 1: "ayin",  2: "bet",  3: "dalet",  4: "gimel",  5: "he",  6: "het", 
                     7: "kaf",  8: "kaf-final",  9: "lamed",  10: "mem",  11: "mem-medial",  12: "nun-final", 
                     13: "nun-medial",  14: "pe",  15: "pe-final",  16: "qof",  17: "resh",  18: "samekh",  19: "shin", 
                     20: "taw",  21: "tet",  22: "tsadi-final",  23: "tsadi-medial",  24: "waw",  25: "yod",  26: "zayin"
            }
    for i in range(len(x_test)):
        print(i)
        image = x_test[i]
        #plt.figure(figsize = (50,5))
        #plt.imshow(image, cmap='gray', aspect = 1)
        #plt.show()
        image = np.reshape(image, (1,100,100))
        #print(y_test[test_idx])
        preds = model.predict(image, batch_size=None, verbose=0, steps=None)
        mean = np.mean(preds)
        std = np.std(preds)
        #print(preds)
        #preds = np.where(preds >= 0.5)
        #preds = preds[1]
        
        
        answer = []
        for j in range(len(y_test[i])):
            if(y_test[i][j] == 1):
                answer.append(labels[j])
        
        preds_sorted = np.sort(preds)
        #print("Preds:", preds)
        preds_values = preds_sorted[:, 27 - len(answer):]
        #print(preds_values)
        #print("Preds:", preds)
        mn = np.min(preds_values)
        preds = np.where(preds >= mn)
        preds = preds[1]
        
        #print("preds",preds)
        pred = np.zeros((27,1))

        for j in preds:
            pred[j,:] = 1
        

        

       # print(answer)

        predic = []
        #print("Prediction:")
        for j in range(len(pred)):
            if(pred[j] == 1):
                predic.append(labels[j])
        
        if(set(answer) == set(predic)):
             correct += 1
        #print(answer)
        #print(predic)
        #print("----------------------------------------")
    print("Correct:", correct)
    print("Total:", len(x_test))
    print("% correct:", correct / len(x_test))
```

```python
test_lstm()
```

```python
def save_model(model):
    json_string = model.to_json()
    model.save_weights("stored_weights")
    
    with open("model.json", "w") as json_file:
        json_file.write(json_string)
    
def load_model():   
    json_file = open('best_model/model.json', 'r')
    loaded_model_json = json_file.read()
    
    model = model_from_json(loaded_model_json)
    model.load_weights("best_model/stored_weights")
    return model
```
