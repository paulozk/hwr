```python
def learn_key(key, value):
    if key not in memory:
        memory[key] = []
    memory[key].append(value)
```

```python
def store_bigrams(names):
    for i in range(len(names)):
        word = names[i]
        char_in_word = word.split('_')

        # cbigrams_listreate a list of bigrams from each word
        bigrams_in_word = [[char_in_word[j], char_in_word[j+1]] for j in range(len(char_in_word)-1)]
        for idx in range(len(bigrams_in_word)):
            bigram = bigrams_in_word[idx]
            learn_key(bigram[0], bigram[1])
```

```python
def get_probability(memory):
    # calculate the frequency of next chars after a certain char
    for key, values in memory.items():
        d = defaultdict(int)
        for value in values:
            d[value] += 1
        total_freqs = len(values)
#         print(key)
#         print(d)
#         print(total_freqs) 

        # calculate the probability of each char after a certain char
        f = defaultdict(float)
        for value in values:
            f[value] = d[value]/total_freqs
        print("key :", key)
        print(f)
```
